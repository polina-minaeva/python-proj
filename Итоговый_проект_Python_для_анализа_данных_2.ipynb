7) Проверим гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

```python
df_two_par = df_survey[['average_montly_hours', 'salary']]
```

```python
df_low = df_two_par[df_two_par['salary']== 'low']
```

```python
list_low = df_low['average_montly_hours'].tolist() #создадим два списка в которые поместим згначения часов работы
                                                    #для зарплат - низких и высоких
```

```python
df_high = df_two_par[df_two_par['salary']== 'high']
```

```python
list_high = df_high['average_montly_hours'].tolist()
```

```python
print(list_low)
```

```python
"""Проверим гипотезу, что сотрудники с высоким окладом
проводят на работе больше времени, чем сотрудники с низким
окладом"""
import scipy.stats as st

stat, p = st.ttest_ind(list_high, list_low)

print(f"Статистика = {stat:.5f}, p = {p:.10f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, средние, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, средние, вероятно, различаются')
```

Так как мы не отклюняем нулевую гипотезу, можно предположить, что сотрудники с высоким окладом не проводят больше времени на работе в статистически значимом смысле, чем сотрудники с низким окладом

9) Разделим данные на тестовую и обучающую выборки. Построим модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary). Также оценим качество модели на тестовой выборке

```python
df_survey_x = df_survey[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'left', 'promotion_last_5years']]
```

```python
df_survey_x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_survey_x[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years']]
y = df_survey_x['left']
```

```python
import numpy as np 
y = np.array(y)
```

```python
#приведем данные к единому масштабу
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
# обучаем данные
lda.fit(X_train, y_train)
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

```python
# смотрим разницу факта и прогноза
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
```

```python
result.head(30)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lda.predict(X_test))
#проверяем точность прогноза
```

С помощью модели мы предсказали увольнение сотрудников компании с точностью более 75%
